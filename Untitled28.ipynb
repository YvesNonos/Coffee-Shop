Here's a breakdown of the columns in your dataset:

**Identifiers and Demographics:**

*   **`ID`**: Unique customer identifier.
*   **`Year_Birth`**: Year of birth of the customer.
*   **`Education`**: Customer's education level (e.g., Graduation, PhD, Master).
*   **`Marital_Status`**: Customer's marital status (e.g., Single, Together, Married).
*   **`Income`**: Customer's yearly household income (note: there are some missing values).
*   **`Kidhome`**: Number of small children in the customer's household.
*   **`Teenhome`**: Number of teenagers in the customer's household.
*   **`Dt_Customer`**: Date of customer's enrollment with the company.
*   **`Recency`**: Number of days since the last purchase.

**Product Spend (Amount spent in the last 2 years on various products):**

*   **`MntWines`**: Amount spent on wine products.
*   **`MntFruits`**: Amount spent on fruit products.
*   **`MntMeatProducts`**: Amount spent on meat products.
*   **`MntFishProducts`**: Amount spent on fish products.
*   **`MntSweetProducts`**: Amount spent on sweet products.
*   **`MntGoldProds`**: Amount spent on gold products.

**Purchase Channels:**

*   **`NumDealsPurchases`**: Number of purchases made with a discount.
*   **`NumWebPurchases`**: Number of purchases made through the companyâ€™s website.
*   **`NumCatalogPurchases`**: Number of purchases made using a catalog.
*   **`NumStorePurchases`**: Number of purchases made directly in stores.
*   **`NumWebVisitsMonth`**: Number of visits to the companyâ€™s website in the last month.

**Campaign Performance:**

*   **`AcceptedCmp1`**: 1 if the customer accepted the offer in the 1st campaign, 0 otherwise.
*   **`AcceptedCmp2`**: 1 if the customer accepted the offer in the 2nd campaign, 0 otherwise.
*   **`AcceptedCmp3`**: 1 if the customer accepted the offer in the 3rd campaign, 0 otherwise.
*   **`AcceptedCmp4`**: 1 if the customer accepted the offer in the 4th campaign, 0 otherwise.
*   **`AcceptedCmp5`**: 1 if the customer accepted the offer in the 5th campaign, 0 otherwise.
*   **`Response`**: 1 if the customer accepted the offer in the last campaign, 0 otherwise.

**Other:**

*   **`Complain`**: 1 if the customer complained in the last 2 years, 0 otherwise.
*   **`Z_CostContact`**: Cost of contact (constant value, likely a categorical or irrelevant feature).
*   **`Z_Revenue`**: Revenue (constant value, likely a categorical or irrelevant feature).

```python
# Calculate the median for each columns
medians = numeric_df.median()

# Replace outliers with the median of each variables
for column in numeric_df.columns:
  Q1 = numeric_df[column].quantile(0.25)
  Q3 = numeric_df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  numeric_df[column] = numeric_df[column].apply(lambda x: medians[column] if x < lower_bound or x > upper_bound else x)
```

```python
# Plot thebox plot after handling the outliers
plt.figure(figsize = (10, 6))
sns.boxplot(data = numeric_df)
plt.xticks(rotation = 45)
plt.title('Box Plot of Numeric Columns')
plt.show()
```
