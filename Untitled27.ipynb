```python
# Rename columns
loan_data = loan_data.rename(columns = {"ApplicantIncome": "Applicant_Income", "CoapplicantIncome": "Coapplicant_Income",
                                "LoanAmount": "Loan_Amount", })
loan_data.head()
```

```python
# Separate categorical features from numerical features
categorical_features = [col for col in loan_data.columns if loan_data[col].dtype == 'object']
numeric_features = [col for col in loan_data.columns if loan_data[col].dtype != 'object']
```

```python
print(f"categorical_features: {categorical_features}")
print(f"numeric_features: {numeric_features}")
```

```python
# Separate categorical from numerical features
categorical_features =  loan_data.select_dtypes(include = 'object')
numeric_features = loan_data.select_dtypes(exclude = 'object')
```

```python
# Handling missing values
loan_data['Gender'].fillna(loan_data['Gender'].mode()[0], inplace = True)
loan_data['Married'].fillna(loan_data['Married'].mode()[0], inplace = True)
loan_data['Dependents'].fillna(loan_data['Dependents'].mode()[0], inplace = True)
loan_data['Education'].fillna(loan_data['Education'].mode()[0], inplace = True)
loan_data['Self_Employed'].fillna(loan_data['Self_Employed'].mode()[0], inplace = True)
loan_data['Applicant_Income'].fillna(loan_data['Applicant_Income'].mean(), inplace = True)
loan_data['Coapplicant_Income'].fillna(loan_data['Coapplicant_Income'].mean(), inplace = True)
loan_data['Loan_Amount'].fillna(loan_data['Loan_Amount'].mean(), inplace = True)
loan_data['Loan_Amount_Term'].fillna(loan_data['Loan_Amount_Term'].mean(), inplace = True)
loan_data['Credit_History'].fillna(loan_data['Credit_History'].mode()[0], inplace = True)
```

```python
loan_data.isnull().sum()
```

```python
loan_data.dtypes
```

```python
loan_data['Dependents'] = loan_data['Dependents'].astype(str).str.replace('+', '', regex=False)
loan_data['Dependents']
```

```python
# Correct the data type
loan_data['Dependents'] = loan_data['Dependents'].astype(int)
loan_data['Dependents'].dtype
```

```python
# Handling Outliers
numeric_cols = loan_data.select_dtypes(include=np.number) # Select only numeric columns
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 +1.5*IQR
print(f"lower_bond: {lower_bound}")
print(f"upper_bond: {upper_bound}")
```

```python
# Visualizing outliers
sns.scatterplot(x = 'Applicant_Income', y = 'Loan_Amount', data = loan_data)
```

```python
# Plot The Outliers
# Help visualize the outliers for each numerical features
plt.figure(figsize = (9, 6))
numeric_features.boxplot(rot = 45)
plt.tile = 'Outliers'
plt.tight_layout()
plt.show()
```

```python
# Create subplots to visualize outliers with boxplots
n_cols = 3
n_rows = 3
fig, axes = plt.subplots(n_rows, n_cols, figsize = (n_cols*5, n_rows*4))
axes = axes.flatten()

for i, col in enumerate(numeric_features.columns):
  if i < len(axes): # Ensure we don't go out of bounds if there are more columns than subplots
    axes[i].boxplot(numeric_features[col].dropna())
    axes[i].set_title(f"Boxplot of {col}")
    axes[i].set_ylabel(col)

# Hide unused subplots
for j in range(i + 1, len(axes)):
  axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```
