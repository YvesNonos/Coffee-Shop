Data Cleaning 4 Customer Call List

```python
# Import the libraries
import pandas as pd
```

```python
# Load the datsset
df = pd.read_excel("/content/Customer Call List.xlsx")
df
```

```python
# Dorp duplicates
df.duplicated()
```

```python
df = df.drop_duplicates()
df
```

```python
# Drop unnecessary column
df = df.drop(columns = ['Not_Useful_Column'])
df
```

```python
# Get rid of the characters in Last_Name
df['Last_Name'].str.lstrip('')
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('/,...,_')
df
```

```python
df
```

```python
# Format Phone Number
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '-', regex = True)
df['Phone_Number'].fillna('', inplace = True)
df
```

```python
df[['Street_Address', 'State', 'Zip Code']]= df['Address'].str.split(',', n=2, expand = True )
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df = df.fillna('')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('NNN', 'N')
df
```

```python
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] =='Y':
    df.drop(x, inplace = True)

df
```

```python
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace = True)

df
```

```python
df = df.drop(columns = ['Address'])
```

```python
# Reset the index
df = df.reset_index(drop = True)
df
```
