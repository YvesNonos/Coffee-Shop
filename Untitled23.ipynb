```python
# Import the librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load Dataset
hotel_bookings = pd.read_csv("/content/hotel_bookings.csv.zip")
hotel_bookings
```

```python
hotel_bookings.head()
```

```python
hotel_bookings.columns
```

```python
# Drop/Rename Columns
hotel_bookings.drop(columns = ['unamed'], axis = 1, inplace = True)
```

```python
# Drop/Rename Columns
hotel_bookings.rename(columns={'adults': 'num_adults', 'children': 'num_children', 'babies': 'num_babies'}, inplace=True)
hotel_bookings.head(10)
```

```python
# NaNs
hotel_bookings.isnull().sum()*100/len(hotel_bookings)
```

```python
# @title Default title text
hotel_bookings['agent'].unique()
hotel_bookings[hotel_bookings['agent'] == 5]
hotel_bookings['agent'].fillna(-1, inplace = True)
hotel_bookings[hotel_bookings['num_children'].isna()]
hotel_bookings[hotel_bookings['country'].isna()][['hotel', 'is_canceled', 'country']]
hotel_bookings['country'].fillna('unknown', inplace = True)
hotel_bookings.dropna(subset=['num_children'], inplace = True)
```

```python
hotel_bookings.drop(columns = ['company'], inplace = True, errors='ignore')
hotel_bookings.head()
```

```python
# Check column data types
hotel_bookings.dtypes
```

```python
# Correct the data types
hotel_bookings = hotel_bookings.astype({'is_canceled': 'boolean', 'is_repeated_guest': 'boolean', 'num_children': 'int', 'agent': 'int'})
hotel_bookings.dtypes
```

```python
# Bin columns
# Reduce the number of unique values we have into groups
hotel_bookings['lead_time'].unique()
hotel_bookings['lead_time'].describe() # Help grouping the results

bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800']

hotel_bookings['lead_time_binned'] = pd.cut(hotel_bookings['lead_time'], bins=bins, labels=labels)
hotel_bookings[['lead_time', 'lead_time_binned']]
```

```python
# How to seperate the data from one column into multiple columns
hotel_bookings.head()
```

```python
# Seperate the columns
# hotel_bookings['arrival_date_month'] = hotel_bookings['arrival_date'].str.split('-', expand=True)[0] # This column does not exist
# hotel_bookings['arrival_date_year'] = hotel_bookings['arrival_date'].str.split('-', expand=True)[1] # This column does not exist
hotel_bookings.head()

column_to_move = hotel_bookings.pop('arrival_date_month')
hotel_bookings.insert(4, 'arrival_date_month', column_to_move)
hotel_bookings.head()

column_to_move = hotel_bookings.pop('arrival_date_year')
hotel_bookings.insert(5, 'arrival_date_year', column_to_move)
hotel_bookings.head()
```

```python
hotel_bookings['hotel'].unique()
```

```python
# String cleaning
hotel_bookings['hotel'] = hotel_bookings['hotel'].replace(r"[\*\n\^]", '', regex = True)
hotel_bookings['hotel'].unique()
```

```python
# Remove Duplicated
hotel_bookings.loc[hotel_bookings.duplicated(keep = False)]
```

```python
# Remove Duplicates
hotel_bookings.drop_duplicates()
hotel_bookings.loc[:, ~hotel_bookings.columns.duplicated(keep=False)]
```
