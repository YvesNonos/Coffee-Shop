```python
# Question 5, Hundreds, fifties, ducks (0) AVG by country
# Ensure '100', '50', and '0' columns are numeric (handling non-numeric values)
cricket_dataset['100'] = pd.to_numeric(cricket_dataset['100'], errors='coerce')
cricket_dataset['50'] = pd.to_numeric(cricket_dataset['50'], errors='coerce')
cricket_dataset['0'] = pd.to_numeric(cricket_dataset['0'], errors='coerce')

# Calculate the mean of '100', '50', and '0' columns grouped by 'Country'
country_stats = cricket_dataset.groupby('Country')[['100', '50', '0']].mean()

print("Average Hundreds, Fifties, and Ducks by Country:")
print(country_stats)
```
